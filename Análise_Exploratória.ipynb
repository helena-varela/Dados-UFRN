```python
# Importações

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_exemplos_limpo.csv')
```

```python
# Mostre as 5 primeiras linhas do dataset.

print('Tabela:\n')
print(display(df.head()))
```

```python
# Obtenha informações estruturais: número de linhas, colunas e tipos de dados.

print('Linhas e colunas:\n')
print(df.shape)
```

```python
# Gere estatísticas descritivas das colunas numéricas (CRA, Taxa_Mensalidade, Horas_Atividade_Extra).

df.describe()[["CRA", "Taxa_Mensalidade", "Horas_Atividade_Extra"]]
```

```python
# Mostre apenas as colunas ID_Aluno, Nome e CRA.

display(df[["Nome","ID_Aluno", "CRA"]])
```

```python
# Quantos cursos diferentes existem no dataset? Quais são eles?

display(df["Curso"].value_counts())
```

```python
# Qual a proporção de alunos por curso?

df['Curso'].value_counts(normalize=True)
```

```python
# Filtre os alunos com CRA maior que 8. 8.

CRA_Maior_8 = df[df["CRA"] > 8.8]
CRA_Maior_8
```

```python
# Filtre apenas os alunos que possuem bolsa.

df[df['Possui_Bolsa'] == True]
```

```python
# Filtre os alunos do curso Psicologia com CRA acima de 7.5.

psico = df[(df['Curso'] == 'Psicologia') & (df["CRA"] > 7.5)]
psico
```

```python
# Filtre os alunos cuja mensalidade é maior que 2500 reais.

mensalidade = df[df["Taxa_Mensalidade"] > 2500]
mensalidade
```

```python
#  Agrupe os dados por Curso e calcule a média do CRA por curso.

media = df.groupby("Curso")['CRA'].mean().round(3).sort_values(ascending=False)
media
```

```python
# Qual a média de horas de atividade extra por período atual? (agrupado por Periodo_Atual)

media2 = df.groupby("Periodo_Atual")["Horas_Atividade_Extra"].mean().round(2)
media2
```

# **Exercícios de visualização de dados**

```python
# Como os alunos estão distribuídos em cada curso?

alunos_curso = df["Curso"].value_counts()

plt.bar(alunos_curso.index, alunos_curso.values, color='pink')
plt.title("Distribuição de Alunos por Curso")
plt.xlabel("Cursos")
plt.ylabel("Alunos")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Como é distribuída a mensalidade da faculdade de maneira geral?

plt.hist(df["Taxa_Mensalidade"], color='blue', edgecolor='black')
plt.title("Histograma de Distribuição Geral de mensalidade")
plt.xlabel("Mensalidade (R$)")
plt.ylabel("Número de Alunos")
plt.show()
```

```python
# Quais são os valores que variam nas mensalidades para cada curso?

plt.bar(df["Curso"], df["Taxa_Mensalidade"], color='purple')
plt.title("Distribuição Geral da mensalidade")
plt.xlabel("Cursos")
plt.ylabel("Taxa Mensalidade")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Qual foi a progressão da média dos alunos durante os anos?

df_progressao = df.groupby("Ano_Ingresso")['CRA'].mean()
plt.plot(df_progressao.index, df_progressao.values)
plt.title("Média do CRA por Ano de Ingresso")
plt.xlabel("Ano de Ingresso")
plt.ylabel("Média do CRA")
plt.grid(True)
plt.show()
```
